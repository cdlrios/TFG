```r
# análisis correlaciones --------------------------------------------------
matriz_corr <- cor(metro_data)
matriz_corr["demanda",]

altamente_correladas <- findCorrelation(matriz_corr, cutoff = 0.9)
data_reducida <- metro_data[,-altamente_correladas]

variables <- setdiff(names(data_reducida),"demanda")
```
